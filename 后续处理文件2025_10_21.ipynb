```python
# -*- coding: utf-8 -*-
# ======================================================
#  移动副本文件到备份文件夹的脚本
# ======================================================

import os
import shutil

print("开始移动副本文件到备份文件夹...")

# --- 文件夹配置 ---
# 确保这些名称与您的Google Drive中的文件夹名称一致
DRIVE_ROOT = '/content/drive/MyDrive'
SOURCE_FOLDER_NAME = 'S2_700_Staging_Exports'      # 包含副本文件的文件夹
BACKUP_FOLDER_NAME = 'S2_700_Staging_Exports_BACKUP' # 您的备份目标文件夹 (如果名称不同，请在这里修改)

# 构建完整的路径
source_path = os.path.join(DRIVE_ROOT, SOURCE_FOLDER_NAME)
backup_path = os.path.join(DRIVE_ROOT, BACKUP_FOLDER_NAME)

print(f"源文件夹: {source_path}")
print(f"目标备份文件夹: {backup_path}")

# --- 核心移动逻辑 ---
# 1. 确保备份文件夹存在，如果不存在则创建它
os.makedirs(backup_path, exist_ok=True)

moved_count = 0
# 2. 检查源文件夹是否存在
if not os.path.exists(source_path):
    print(f"\n错误：找不到源文件夹 '{source_path}'。请检查名称是否正确。")
else:
    # 3. 遍历源文件夹中的所有文件
    for filename in os.listdir(source_path):
        # 4. 检查文件名是否包含“(副本)”来识别副本文件
        if '（副本）' in filename:
            try:
                # 构建完整的文件路径
                source_file = os.path.join(source_path, filename)
                destination_file = os.path.join(backup_path, filename)

                # 移动文件
                shutil.move(source_file, destination_file)
                print(f"  已移动: {filename}")
                moved_count += 1
            except Exception as e:
                print(f"移动文件 {filename} 时出错: {e}")

    print("\n========================================================")
    if moved_count > 0:
        print(f"移动完成！共 {moved_count} 个副本文件已成功移动到备份文件夹。")
        print("您的原始 'S2_700_Staging_Exports' 文件夹现在是干净的了。")
    else:
        print("操作完成，但在源文件夹中没有找到带有 '(副本)' 标志的文件。")
    print("========================================================")
```

```python
# -*- coding: utf-8 -*-
# ======================================================
#  第三单元格：终极智能修复波段名称脚本 (最终修正版)
# ======================================================

import rasterio
import re # 导入正则表达式库

print("开始运行终极智能修复脚本...")

if not os.path.exists(final_path):
    print(f"错误：找不到最终文件夹 '{final_path}'。请确保文件已整理。")
else:
    repaired_count = 0
    skipped_count = 0
    # 遍历最终文件夹下的所有站点子文件夹
    for site_folder in os.listdir(final_path):
        site_folder_path = os.path.join(final_path, site_folder)
        if os.path.isdir(site_folder_path):
            # 遍历每个站点文件夹里的 .tif 文件
            for filename in os.listdir(site_folder_path):
                if filename.endswith('.tif'):
                    file_path = os.path.join(site_folder_path, filename)
                    try:
                        print(f"\n正在检查文件: {filename}")
                        needs_update = False
                        new_descriptions = []

                        with rasterio.open(file_path, 'r+') as dataset:
                            if not dataset.descriptions:
                                print("  -> 警告: 文件没有任何波段描述，跳过。")
                                continue

                            old_descriptions = dataset.descriptions
                            first_desc_parts = old_descriptions[0].split('_')

                            if not (len(first_desc_parts) == 3 and first_desc_parts[0].isdigit()):
                                needs_update = True
                                print(f"  检测到需要修复的格式: '{old_descriptions[0]}'")
                            else:
                                print(f"  格式已正确 ('{old_descriptions[0]}'), 跳过。")
                                skipped_count += 1
                                continue

                            if needs_update:
                                for old_desc in old_descriptions:
                                    # 1. 提取日期 (YYYYMMDD) - 查找8位数字
                                    date_match = re.search(r'(\d{8})', old_desc)
                                    date_str = date_match.group(1) if date_match else "NODATE"

                                    # 2. 提取波段号 (例如 B4)
                                    band_match = re.search(r'(B\d{1,2}A?)', old_desc)
                                    band_id = band_match.group(1) if band_match else "NOBAND"

                                    # 3.【关键修正】使用严格规则匹配瓦片ID (T + 2位数字 + 3位大写字母)
                                    #    这可以正确忽略时间戳里的'T'
                                    tile_match = re.search(r'(T\d{2}[A-Z]{3})', old_desc)
                                    tile_id = tile_match.group(1) if tile_match else "NOTILE"

                                    # 4. 组合成新的标准格式
                                    new_desc = f"{date_str}_{band_id}_{tile_id}"
                                    new_descriptions.append(new_desc)

                                # 安全检查：如果任何一个部分没找到，就放弃修改这个文件
                                if "NODATE" in new_descriptions[0] or "NOBAND" in new_descriptions[0] or "NOTILE" in new_descriptions[0]:
                                     print(f"  -> 警告: 无法从 '{old_descriptions[0]}' 中完整提取信息，已跳过此文件以保证数据安全。")
                                     continue

                                # 将新的波段描述写回文件
                                dataset.descriptions = tuple(new_descriptions)
                                print(f"  -> 成功将 {len(new_descriptions)} 个波段重命名为新格式 (例如: '{new_descriptions[0]}')")
                                repaired_count += 1

                    except Exception as e:
                        print(f"处理文件 {filename} 时出错: {e}")

    print("\n========================================================")
    print("修复完成！")
    print(f"共修复了 {repaired_count} 个文件。")
    print(f"共跳过了 {skipped_count} 个格式正确的文件。")
    print("========================================================")
```

```python
# -*- coding: utf-8 -*-
# ======================================================
#  第四单元格：外科手术式修复脚本 (针对性修复已知错误)
# ======================================================

import rasterio
import re
import os

print("开始运行针对性的“外科手术式”修复脚本...")

# --- 手动定义的修复地图 ---
# 在这里，我们明确指定哪个站点需要修复，以及它正确的瓦片ID是什么。
# 格式: { '站点文件夹名': '正确的瓦片ID' }
# 您可以根据需要添加更多条目。
manual_fix_map = {
    'norriepoint': 'T18TWM'
}

print("\n--- 将要执行的修复任务 ---")
print(manual_fix_map)
print("脚本将只处理以上列出的站点文件夹。")
# -----------------------------

repaired_count = 0
skipped_count = 0

if not os.path.exists(final_path):
    print(f"\n错误：找不到最终文件夹 '{final_path}'。")
else:
    # 遍历我们手动指定的需要修复的站点
    for site_name, correct_tile_id in manual_fix_map.items():
        site_folder_path = os.path.join(final_path, site_name)

        if not os.path.exists(site_folder_path):
            print(f"\n警告：在修复列表中指定的站点文件夹 '{site_name}' 不存在，已跳过。")
            continue

        print(f"\n--- 正在扫描目标站点: '{site_name}' (强制使用正确瓦片ID: {correct_tile_id}) ---")

        # 遍历目标站点文件夹中的所有TIF文件
        for filename in os.listdir(site_folder_path):
            if filename.endswith('.tif'):
                file_path = os.path.join(site_folder_path, filename)
                try:
                    print(f"  -> 正在检查文件: {filename}")
                    with rasterio.open(file_path, 'r+') as dataset:
                        if not dataset.descriptions:
                            print("     文件无波段描述，跳过。")
                            continue

                        first_desc = dataset.descriptions[0]

                        # 【关键】检查波段名是否是之前被损坏的格式 (T + 5个数字)
                        if re.search(r'T\d{5}', first_desc):
                            print(f"     [发现损伤!] 格式为: '{first_desc}'。正在应用修复...")
                            old_descriptions = dataset.descriptions
                            new_descriptions = []

                            for old_desc in old_descriptions:
                                # 从损坏的名称中提取仍然有用的日期和波段号
                                date_match = re.search(r'(\d{8})', old_desc)
                                date_str = date_match.group(1) if date_match else "NODATE"

                                band_match = re.search(r'(B\d{1,2}A?)', old_desc)
                                band_id = band_match.group(1) if band_match else "NOBAND"

                                # 使用我们手动指定的、100%正确的瓦片ID来组合新名称
                                new_desc = f"{date_str}_{band_id}_{correct_tile_id}"
                                new_descriptions.append(new_desc)

                            # 将修复后的名称写回文件
                            dataset.descriptions = tuple(new_descriptions)
                            print(f"     [修复成功] 已重命名为 (例如): '{new_descriptions[0]}'")
                            repaired_count += 1
                        else:
                            print(f"     格式正确或无需修复: '{first_desc}', 跳过。")
                            skipped_count += 1

                except Exception as e:
                    print(f"处理文件 {filename} 时出错: {e}")

print("\n========================================================")
print("修复完成！")
print(f"共修复了 {repaired_count} 个文件。")
print(f"共跳过了 {skipped_count} 个无需修复的文件。")
print("========================================================")
```
